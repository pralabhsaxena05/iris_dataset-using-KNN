```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
```

```python
#splitting the dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'],iris_dataset['target'], random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#creating dataset from training set
iris_df = pd.DataFrame(X_train, columns = iris_dataset.feature_names)
```

```python
iris_df.head()
```

```python
grr = pd.plotting.scatter_matrix(iris_df,c=y_train,figsize=(15,15),marker='o',hist_kwds={'bins':20},s=60,alpha=0.9)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=2)
```

```python
knn.fit(X_train,y_train)
```

```python
X_new = np.array([[5.9,
3.0,
4.2,
1.5]])
X_new.shape
prediction = knn.predict(X_new)
```

```python
print(prediction)
print(iris_dataset['target_names'][prediction])
```

```python
y_pred = knn.predict(X_test)
y_pred
```

```python
print('{:.2f}'.format(np.mean(y_pred==y_test)))
```

```python
#accuracy
print('Test score is: {:.2f}'.format(knn.score(X_test,y_test)))
```
